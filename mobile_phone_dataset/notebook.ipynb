```python
# went through the unique values in each column and  got some key values in each column that would be useful.
# putting the key values in a dictionary
key_values = {
    "connectivity": ["Wifi", "4G,","3G" ,"3G,", "2G,","5G,", "Sim,", "Wi-Fi", "Sim"],
    "processor" : ["Apple","Bionic", "Processor","Unisoc", "ASR", "Helio"],
    "storage" : "inbuilt",
    "extra_storage" : "Memory",
    "battery" : "Battery",
    "display" : "Display",
    "camera" : "Camera",
    "os" : ["Android", "iOS","HarmonyOS","Symbian","KAI", "Hongmeng", "Flyme", "Nucleus", "KaiOS"],
    "comm" : ["Radio", "Bluetooth"],
}
```

```python
# priting the keys in the dictionary
key_values.keys()
```

```python
# priting the values in the dictionary
key_values.values()
```

```python
# Writing a code to sort out the values from a column and assigning it to a new required column
for column in key_values.keys():  # Iterate over each key in the key_values dictionary
    if column in df.columns:  # Check if the column exists in the DataFrame
        if isinstance(key_values[column], list):  # Check if the value corresponding to the key is a list
            # Create a new column with the cleaned values
            df[f"{column}_cleaned"] = df[column].apply(lambda x: x if any(i in str(x).split() for i in key_values[column]) else np.nan)
            # Update the original column by setting to NaN the values that match any element in the list
            df[column] = df[column].apply(lambda x: np.nan if any(i in str(x).split() for i in key_values[column]) else x)
        else:  # If the value corresponding to the key is not a list
            # Create a new column with the cleaned values
            df[f"{column}_cleaned"] = df[column].apply(lambda x: x if key_values[column] in str(x).split() else np.nan)
            # Update the original column by setting to NaN the values that match the key's value
            df[column] = df[column].apply(lambda x: np.nan if key_values[column] in str(x).split() else x)
    else:  # If the column does not exist in the DataFrame, create a new cleaned column with NaN values
        df[f"{column}_cleaned"] = np.nan
```

```python
# checking the number of nan values in each column to verify if the code worked
df.isnull().sum()
```

```python
# Noticed that the data is not fully cleaned
```

```python
# Writing a logic to fully clean the data
for column in df.columns:  # Iterate over each column in the DataFrame
    if column in key_values.keys():  # Check if the column is in the key_values dictionary
        for i in range(len(df)):  # Iterate over each row in the DataFrame
            for j in key_values.keys():  # Iterate over each key in key_values
                if isinstance(key_values[j], list):  # Check if the value corresponding to the key is a list
                    # Check if any element of the list is in the split string of the column's value in the current row
                    if any(k in str(df[column].iloc[i]).split() for k in key_values[j]):
                        # If the cleaned column is NaN, update it with the current column's value and set the current column's value to NaN
                        if pd.isna(df[f"{j}_cleaned"].iloc[i]):
                            df.loc[i, f"{j}_cleaned"] = str(df[column].iloc[i])
                            df.loc[i, column] = np.nan
                        else:  # If the cleaned column is not NaN, append the current column's value to the cleaned column
                            df.loc[i, f"{j}_cleaned"] = str(df[f"{j}_cleaned"].iloc[i]) + ", " + str(df[column].iloc[i])
                            df.loc[i, column] = np.nan
                else:  # If the value corresponding to the key is not a list
                    # Check if the value is in the split string of the column's value in the current row
                    if key_values[j] in str(df[column].iloc[i]).split():
                        # If the cleaned column is NaN, update it with the current column's value and set the current column's value to NaN
                        if pd.isna(df[f"{j}_cleaned"].iloc[i]):
                            df.loc[i, f"{j}_cleaned"] = str(df[column].iloc[i])
                            df.loc[i, column] = np.nan
                        else:  # If the cleaned column is not NaN, append the current column's value to the cleaned column
                            df.loc[i, f"{j}_cleaned"] = str(df[f"{j}_cleaned"].iloc[i]) + ", " + str(df[column].iloc[i])
                            df.loc[i, column] = np.nan
```

```python
# checking the number of nan values in each column to verify if the logic worked
df.isnull().sum()
```

```python
# writing a function to check for values that were in the wrong columns
def check_uncleaned(col):
    for i in df[col]:
        if pd.isna(i) != True:
            print(i)
```

```python
# checking for values that are in each wrong columns 
for i in key_values.keys():
    try:
        print("*"*20)
        print(f"{i}:")
        check_uncleaned(i)
        print("*"*20)
        print("\n")
    except:
        print("*"*20)
```
